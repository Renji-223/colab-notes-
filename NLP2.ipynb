```python
import nltk
```

```python
from nltk.stem.snowball import stopwords
nltk.download('stopwords')
sw=stopwords.words('english')
print(sw)
```

```python
sent='Football worldcup fifa in every five year'
nltk.download('punkt')
tokens=nltk.word_tokenize(sent)
tokens
```

```python
sw=set(stopwords.words('english'))
lower_tokens=list(map(lambda x:x.lower(),tokens))
lower_tokens
```

```python
sw=set(stopwords.words('english'))
fill_token=[w for w in lower_tokens if not w.lower() in sw]
fill_token
```
