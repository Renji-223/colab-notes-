```python
import pandas as pd
df=pd.read_csv('/content/diabetes.csv')
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
X=df.iloc[0:,:-1]
X
```

```python
y=df.iloc[0:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
X_train_new=sc.transform(X_train)
print(X_train_new)
```

```python
X_test_new=sc.transform(X_test)
print(X_test_new)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_new,y_train)
y_pred_knn=knn.predict(X_test_new)
y_pred_knn
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred_knn))
print(classification_report(y_test,y_pred_knn))
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train_new,y_train)
y_pred_nb=nb.predict(X_test_new)
print(y_pred_nb)
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred_nb))
```

```python
print(classification_report(y_test,y_pred_nb))
```

```python
#SVM-SUPPORT VECTOR CLASSIFIER
from sklearn.svm import SVC 
sv=SVC(kernel='linear')
sv.fit(X_train_new,y_train)
y_pred_sv=sv.predict(X_test_new)
print(y_pred_sv)
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred_sv))
print(classification_report(y_test,y_pred_sv))
```
