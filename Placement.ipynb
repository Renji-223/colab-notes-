```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
cols=['gender','ssc_b','hsc_b','hsc_s','degree_t','workex','specialisation','status']
for i in cols:
   df[i]=encoder.fit_transform(df[i])
df
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.feature_selection import VarianceThreshold
select_features=VarianceThreshold(threshold=4.0)
vec=select_features.fit_transform(X,y)
vec=select_features.get_support()
vec
```

```python
col=X.columns
X_variance_threshold=pd.DataFrame({'feature':col,'threshold':vec})
X_variance_threshold
```

```python
X=df[['ssc_p','hsc_p','degree_p','etest_p','mba_p']]
```

```python
X.head()
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python

from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
sc.fit(X_train)
X_train_new=sc.transform(X_train)
X_train_new
```

```python

X_test_new=sc.transform(X_test)
X_test_new
```

```python

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train_new,y_train)
y_pred=knn.predict(X_test_new)
y_pred
```

```python

from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(X_train_new,y_train)
y_pred=reg.predict(X_test_new)
y_pred
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```

```python
#Naive base
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train_new,y_train)
y_pred=nb.predict(X_test_new)
y_pred
```

```python

from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```

```python

from sklearn.svm import SVC 
clf=SVC()
clf.fit(X_train_new,y_train)
y_pred=clf.predict(X_test_new)
y_pred
```

```python

from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
